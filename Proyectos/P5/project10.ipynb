```python
#### PPT 3 DE LA DIAPOSITIVA 24 --- EJEMPLO DE PROGRAMACION DE UN AFD

while True:
    cadena = input("\n\nIngrese una cadena de 0s y 1s : ")
    inicio = 1
    aceptado = 1
    estado = inicio
    fin = False
    contador = 0

    while not fin:
        if contador > len(cadena) - 1:
            fin = True
            break
        if estado == 1:
            if cadena[contador] == '1':
                estado = 1
            if cadena[contador] == '0':
                estado = 2
            contador += 1
            continue
        if estado == 2:
            if cadena[contador] == '1':
                estado = 2
            if cadena[contador] == '0':
                estado = 1
            contador += 1
            continue

    if estado == aceptado:
        print("La cadena", cadena, "es aceptada")
    else:
        print("La cadena", cadena, "es rechazada")
```

```python
#### PPT 3 DE LA DIAPOSITIVA 26 --- EL PROGRAMA DETERMINA SI EL NUMERO TIENE PARTE DECIMAL
print("El programa determina si el numero tiene parte decimal: ")
print("El alfabeto aceptado esta formado por: +,-,.,0-9")

while True:
    input_str = input("\nIngrese un numero: ")
    inicio = 'A'
    final = 'E'
    actual = inicio
    fin = False
    contador = 0

    while not fin:
        if contador > len(input_str) - 1:
            fin = True
            break

        if actual == 'A':
            if input_str[contador].isdigit():
                actual = 'C'
            if input_str[contador] in ['-', '+']:
                actual = 'B'
            if input_str[contador] == '.':
                actual = 'F'
            contador += 1
            continue

        if actual == 'B':
            if input_str[contador].isdigit():
                actual = 'C'
            if input_str[contador] in ['.', '+', '-']:
                actual = 'F'
            contador += 1
            continue

        if actual == 'C':
            if input_str[contador].isdigit():
                actual = 'C'
            if input_str[contador] == '.':
                actual = 'D'
            if input_str[contador] in ['+', '-']:
                actual = 'F'
            contador += 1
            continue

        if actual == 'D':
            if input_str[contador].isdigit():
                actual = 'E'
            if input_str[contador] in ['+', '-']:
                actual = 'F'
            contador += 1
            continue

        if actual == 'E':
            if input_str[contador].isdigit():
                actual = 'E'
            if input_str[contador] in ['.', '+', '-']:
                actual = 'F'
            contador += 1
            continue

        if actual == 'F':
            if input_str[contador].isdigit():
                actual = 'F'
            if input_str[contador] in ['.', '+', '-']:
                actual = 'F'
            contador += 1
            continue

    if actual == final:
        print("La cadena", input_str, "es REAL")
    else:
        print("La cadena", input_str, "es rechazada")
        print("Estado final:", actual)
```

```python
### EJERCCIO PARA CASA PPT 3 DIAPOSITIVA 31
def es_correo_valido(cadena):
    estado = 'A'  # Estado inicial
    for caracter in cadena:
        if estado == 'A':
            if caracter.isalnum() or caracter in ['.', '_', '-', '+']:
                estado = 'A'
            elif caracter == '@':
                estado = 'B'
            else:
                estado = 'C'
                break
        elif estado == 'B':
            if caracter.isalnum() or caracter in ['.', '-']:
                estado = 'B'
            else:
                estado = 'C'
                break
    return estado == 'B'


cadena = input("Ingrese una dirección de correo electrónico: ")
if es_correo_valido(cadena):
    print("Aceptado")
else:
    print("Rechazado")
```
